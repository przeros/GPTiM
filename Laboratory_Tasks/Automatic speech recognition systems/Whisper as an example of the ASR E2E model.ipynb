[Whisper]( https://openai.com/research/whisper) to zaawansowany system rozpoznawania mowy stworzony przez OpenAI. System ten jest zaprojektowany tak, aby działać skutecznie w różnych językach i dialektach, co stanowi jeden z jego kluczowych atutów. Dodatkową zaletą systemu jest poprawne działanie środowisku mocno zaszumionym. Oznacza to, że system do poprawnej transkrypcji nie wymaga wysokich wartości współczynnika sygnału do szumu (SNR).

Architektura Whisper to proste podejście typu end-to-end, zaimplementowane jako transformer typu koder-dekoder. Dźwięk wejściowy jest dzielony na 30-sekundowe  fragmenty, konwertowane na melspektrogram, a następnie przekazywane do kodera. Dekoder jest trenowany w celu przewidywania odpowiednich podpisów tekstowych, mieszanych ze specjalnymi tokenami, które kierują pojedynczy model do wykonywania konkretnego zadania, takiego jak identyfikacja języka, przypisywania znaczniki czasu na poziomie frazy, wielojęzyczna transkrypcja mowy czy tłumaczenie mowy na język angielski. Whsiper trenowany był na zbiorze 680 tys. godzin nagrań, co obecnie czyni go modelem audio z największym zbiorem treningowym.

Poniżej, w punktach przedstawiono kluczowe cechy architektury Whisper:

1.   Wielojęzyczność: System jest zaprojektowany do
rozpoznawania i transkrypcji mowy w wielu językach

2.   Dokładność transkrypcji: Dzięki zaawansowanym algorytmom i dużemu zbiorowi danych treningowych, Whisper osiąga wysoką dokładność w transkrypcji mowy. Dla jeżyka angielskiego deklarowany WER wynosi 9.3 dla modelu large-v3 i zbioru Common Voice 15. Dla języka polskiego na tym samym zbiorze i tym samym modelu Whisper osiąga WER równy 6. [Tabela dla large-v2](https://www.assemblyai.com/blog/content/images/2022/09/multilingual_wer.png)

3.   Adaptacja do warunków akustycznych: System jest w stanie dostosować się do różnych warunków akustycznych, takich jak hałas w tle, co jest istotne w praktycznym zastosowaniu.

4.   Rozpoznawanie kontekstu: Whisper nie tylko transkrybuje mowę, ale także jest w stanie rozpoznać kontekst wypowiedzi, co poprawia jakość i zrozumiałość transkrypcji – zastosowanie transformera i mechanizmu atencyjnego.

5.   Elastyczność zastosowań: Możliwość wykorzystania w różnych aplikacjach, od asystentów głosowych po systemy automatycznego podtytułowania.

Whisper jest obecnie State of the Art w przypadku systemów ASR. Jego pojawienie się było istotnym krokiem naprzód w dziedzinie rozpoznawania mowy, głównie ze względu na wysoką dokładność, elastyczność oraz wielojęzyczność, co czyni go narzędziem przydatnym w wielu różnych kontekstach i aplikacjach.

Moduł pozwalający na rejstrację sygnału mowy bezpośrednio z przeglądarki

````python
# Przechytywanie strumienia audio przez przeglądarkę
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def record_audio(duration=5):
    js = Javascript("""
    async function recordAudio(duration) {
        const stream = await navigator.mediaDevices.getUserMedia({audio: true});
        const mediaRecorder = new MediaRecorder(stream);
        let audioChunks = [];
        mediaRecorder.ondataavailable = e => audioChunks.push(e.data);
        mediaRecorder.start();
        await new Promise(resolve => setTimeout(resolve, duration * 1000));
        mediaRecorder.stop();
        return new Promise(resolve => {
            mediaRecorder.onstop = () => {
                const audioBlob = new Blob(audioChunks);
                const audioUrl = URL.createObjectURL(audioBlob);
                const audio = new Audio(audioUrl);
                const reader = new FileReader();
                reader.readAsDataURL(audioBlob);
                reader.onloadend = () => {
                    const base64data = reader.result;
                    resolve(base64data);
                };
            };
        });
    }
    """)
    display(js)
    audio = eval_js('recordAudio({})'.format(duration))
    audio_file = open('audio_sample.webm', 'wb')
    audio_file.write(b64decode(audio.split(',')[1]))
    audio_file.close()
    return 'audio_sample.webm'

audio_path = record_audio(duration=5)
ipd.Audio(audio_path)
````

Moduł konwertujący audio zapisane w formacie .webm na .wav, czytelny dla modelu

````python
def convert_webm_to_wav(input_webm, output_wav):
    try:
        # Konwersja z webm z użyciem ffmpeg
        stream = ffmpeg.input(input_webm)
        stream = ffmpeg.output(stream, output_wav)
        ffmpeg.run(stream)
        print(f"File converted successfully: {output_wav}")
    except ffmpeg.Error as e:
        print("Error:", e)

# Przykład użycia
input_webm = audio_path
output_wav = 'audio_sample.wav'
convert_webm_to_wav(input_webm, output_wav)
````

Właściwe ładowanie modelu Whisper i transkrypcja. Nalezy pamiętać, że Whisper przyjmuje pliki audio z częstotliwością próbkowania 16 kHz i posiada własną, wbudowaną parametryzację. Na wyjściu tej komórki otrzymany jest wynik transkrypcji w surowej postaci.

````python
model = whisper.load_model("base")

# ładowanie audio i ręczne dokonanie padowania/trimowania aby miało 30 sekund
audio = whisper.load_audio("audio_sample.wav")
audio = whisper.pad_or_trim(audio)

# stworznie melspektrogramu i  uploadowanie go do GPU
mel = whisper.log_mel_spectrogram(audio).to(model.device)

# Detekcja języka zawartego w pliku audio
_, probs = model.detect_language(mel)
print(f"Detected language: {max(probs, key=probs.get)}")

# Dekodowanie audio
options = whisper.DecodingOptions()
result = whisper.decode(model, mel, options)
# Transkrypcja
transcription = result.text
print(transcription)
````

Teraz należy dokonać normalizacji tekstu, aby popranwie liczyć miarę WER. Do tego zadania potrzebna jest zdefiniowana poniżej funkcja `normalize_text`, która dokonuje normalizacji tekstu porzez wykonanie lowercase'a oraz lemmatyzacji. Poniżej przykład wywołania funkcji, aby porównać tekst przed i po procesie normalizacji:


```
sample_text = "Cats running were faster than dogs."
normalized_sample = normalize_text(sample_text)
print("Original Text:", sample_text)
print("Normalized Text:", normalized_sample)
```



````python
def normalize_text(text):
    text = text.lower()

    # Lemmatyzacja
    lemmatizer = WordNetLemmatizer()
    words = text.split()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    normalized_text = ' '.join(lemmatized_words)

    return normalized_text

# Normalizacja
normalized_manual_text = normalize_text(text)
normalized_transcription = normalize_text(transcription)

# Obliczanie WER dla nieznormalizowanego tekstu
non_normalized_metrics = compute_measures(text, transcription)
print("Non-Normalized Text WER:", non_normalized_metrics['wer'])

# Obliczanie WER dla znormalizowanego tekstu
normalized_metrics = compute_measures(normalized_manual_text, normalized_transcription)
print("Normalized Text WER:", normalized_metrics['wer'])

print("Original Text:", text)
print("Normalized Text:", normalized_manual_text)
print("Original Transcription:", transcription)
print("Normalized Transcription:", normalized_transcription )
````

W tej komórce dokonywana jest właściwa kalkulacja miary WER oraz podawana liczba wstawień, usunięć i podmian. Im mniejsze wartości tym lepsza transkrypcja.

````python
metrics = compute_measures(text, transcription)
print("Word Error Rate (WER):", metrics['wer'])
print("Number of Insertions:", metrics['insertions'])
print("Number of Deletions:", metrics['deletions'])
print("Number of Substitutions:", metrics['substitutions'])
````
